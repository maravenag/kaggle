```python
import pandas as pd
```

```python
data = pd.read_csv("data/BreadBasket_DMS.csv")
```

```python
data.head()
```

```python
## convert date and time into Datetime
data['Timestamp'] = data['Date']+' '+data['Time']
```

```python
data.Timestamp = pd.to_datetime(data.Timestamp)
data.drop(labels=["Date", "Time"], axis=1, inplace=True)
```

```python
items = data.Item.unique().tolist()
```

```python
%matplotlib inline
## plot a few items
for item in items:
    aux = data[data['Item'] == item]
    aux.plot(x="Timestamp", y="Transaction")
```
